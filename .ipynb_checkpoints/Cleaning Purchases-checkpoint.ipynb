```python
import pandas as pd
import re
from pandas import ExcelWriter
```

```python
df = pd.read_excel(open('Prueba Tableau limpia.xlsx', 'rb'),
              sheet_name='Purchases')
df.head()
```

```python
df['Stock ID & Marca'] = df['Stock ID & Marca'].str.replace('\d+', '')
```

```python
df.head()
```

```python
df['Marca'] = df['Stock ID & Marca']
```

```python
df.head()
```

```python
df.drop(['Stock ID & Marca'],axis=1,inplace = True)
```

```python
df.head()
```

```python
df2 = pd.read_excel(open('Prueba Tableau limpia.xlsx', 'rb'),
              sheet_name='Monthly Sales')
df2.head()
```

```python
df3 = pd.read_excel(open('Prueba Tableau limpia.xlsx', 'rb'),
              sheet_name='Initial Inventory')
df3.head()
```

```python
writer = pd.ExcelWriter('Prueba Tableau limpia.xlsx', engine='xlsxwriter')


df.to_excel(writer, sheet_name='Purchases',index = False)
df2.to_excel(writer, sheet_name='Monthly Sales',index = False)
df3.to_excel(writer, sheet_name='Initial Inventory', index = False)

writer.save()
```
