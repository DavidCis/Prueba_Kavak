```python
def save_xls(dict_df, path):

    writer = ExcelWriter(path)
    for key in dict_df:
        dict_df[key].to_excel(writer, key)

    writer.save()
```

```python
df2 = pd.read_excel(open('Prueba Tableau.xlsx', 'rb'),
              sheet_name='Purchases',names=['Stock ID & Marca', 'Precio de Venta Con IVA', '# autos',
                                                       'Tipo de oferta','Fecha de compra','Location Compra',
                                                        'Precio de compra SIN IVA','IVA Precio de Compra',
                                                        'Precio de compra con IVA'])
```

```python
df2
```

```python
df2 = df2.iloc[1:]
df2 = df2.reset_index(drop=True)
```

```python
df2
```

```python
df2[['Stock ID','Marca']] = df2['Stock ID & Marca'].str.split(n=2,expand=True)
```

```python
df2.drop(['Stock ID','Marca'], axis=1,inplace=True)
```

```python
df2
```

```python
s = df2['Stock ID & Marca']
s = s.str.findall(r'[0-9]+')
```

```python
df2['Stock ID'] = s
```

```python
df2.info()
```

```python
text = []
for i in s:
    text.append(i)
```

```python
r = []
for i in text:
    for e in i:
        if e.isdigit():
            e.split()
            r.append(int(e))
```

```python
df2['Stock ID'] = r
```
